```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# 任务 1：分析排名前100项目的 value 数据
openrank_path = r'C:\Users\86189\Desktop\open_rank_20236.csv'
openrank_data = pd.read_csv(openrank_path)

# 选择前100项目
top100 = openrank_data.head(100)

# 计算统计指标
max_value = top100['value'].max()
min_value = top100['value'].min()
mean_value = top100['value'].mean()
median_value = top100['value'].median()

print("前100项目的 value 统计：")
print(f"最大值: {max_value}, 最小值: {min_value}, 均值: {mean_value}, 中位数: {median_value}")

```

```python
# 任务 4：贝叶斯定理计算
P_HTML_non_software = 0.85
P_non_software = 0.25
P_HTML_software = 0.10
P_software = 0.75
P_HTML = P_HTML_non_software * P_non_software + P_HTML_software * P_software
P_non_software_given_HTML = (P_HTML_non_software * P_non_software) / P_HTML
print(f"已知项目带 HTML/Markdown 标签，该项目为非软件型项目的概率: {P_non_software_given_HTML:.2f}")

P_JS_tool = 0.35
P_tool = 0.50
P_JS_non_tool = 0.10
P_non_tool = 0.50
P_JS = P_JS_tool * P_tool + P_JS_non_tool * P_non_tool
P_tool_given_JS = (P_JS_tool * P_tool) / P_JS
print(f"已知项目由 JavaScript 编写，它是工具组件型项目的概率: {P_tool_given_JS:.2f}")
```

```python
# 任务 5：构建可视化无向图
user = [1, 2, 3, 4]
edge = [(1, 2), (2, 3), (3, 4), (4, 1)]
G_undirected = nx.Graph()
G_undirected.add_nodes_from(user)
G_undirected.add_edges_from(edge)

plt.figure(figsize=(8, 6))
nx.draw(G_undirected, with_labels=True, node_size=700, font_weight='bold')
plt.title("无向图")
plt.show()
```
